## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
df.ConvertedComp.plot(kind='box')
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
df.ConvertedComp.describe()
# q1,q3 = np.percentile(df['ConvertedComp'],[25,75])

# iqr = q3 -q1
# iqr
iqr = df.ConvertedComp.describe()[6] - df.ConvertedComp.describe()[4]
iqr
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
df[(df['ConvertedComp'] > upper_bound) | (df['ConvertedComp'] < lower_bound)].shape
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_bound = df.ConvertedComp.describe()[4] - (1.5 * iqr) # Q1 - 1.5*iqr
upper_bound = df.ConvertedComp.describe()[6] + (1.5 * iqr) #Q3 + 15 * iqr

lower_bound, upper_bound

```

```python
# your code goes here
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_outliers_remove = df[df['ConvertedComp'] <= upper_bound]
df_outliers_remove['ConvertedComp'].describe()
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.


```python
# your code goes here
numerical = ['int16','int32','int64','float16','float32','float64']
df_numerical = df.select_dtypes(include=numerical)

Age = df_numerical['Age']
Respondent = df_numerical['Respondent']
Code_Rev_Hrs = df_numerical['CodeRevHrs']
Comp_Total = df_numerical['CompTotal']
Work_Week_Hrs = df_numerical['WorkWeekHrs']
Converted_Comp = df_numerical['ConvertedComp']

Age_respondent_corr = Age.corr(Respondent)
Age_codeRevHrs_corr = Age.corr(Code_Rev_Hrs)
Age_comp_total_corr = Age.corr(Comp_Total)
Age_work_week_hrs = Age.corr(Work_Week_Hrs)
Age_converted_comp = Age.corr(Converted_Comp)

print("The correlation of Age and Respondent is " ,  Age_respondent_corr)
print("The correlation of Age and CodeRevHrs is " , Age_codeRevHrs_corr)
print("The correlation of Age and CompTotal is " , Age_comp_total_corr)
print("The correlation of Age and WorkWeekHrs is " , Age_work_week_hrs)
print("The correlation of Age and ConvertedComp is " , Age_converted_comp)


df_outliers_remove.corr()

```
