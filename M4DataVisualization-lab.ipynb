## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM master
"""
ConvertedComp = pd.read_sql_query(QUERY,conn)

ConvertedComp = ConvertedComp.dropna()
ConvertedComp.isnull().sum()

plt.hist(ConvertedComp, bins=5)
plt.title('Histogram of ConvertedComp')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
Query = """
SELECT Age FROM
master
"""

Age = pd.read_sql_query(Query,conn)
Age.plot(kind='box')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
Query = """
SELECT Age, WorkWeekHrs FROM
master
"""
df_Age_WorkWeekHrs = pd.read_sql_query(Query,conn)


df_Age_WorkWeekHrs.plot(kind = 'scatter', x='Age', y='WorkWeekHrs', figsize=(10,6), color = 'darkblue')

plt.title('Scatter plot of Age and Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
Query = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM
master
"""
df_Bubble = pd.read_sql_query(Query,conn)

# # df_Bubble
norm_Age = (df_Bubble['Age'] - df_Bubble['Age'].min()) / (df_Bubble['Age'].max() - df_Bubble['Age'].min())

df_Bubble.plot(kind='scatter',x='WorkWeekHrs',y='CodeRevHrs', s=norm_Age*1000 + 10, alpha=0.5,color='green', figsize=(14,8))
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs using Age as the size')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.legend(['Age'], loc='upper right', fontsize='x-large')
plt.show()




```
